```python
import numpy as np
from math import gamma
```

```python

def shannon(a):
    a1 = np.array(a)
    b=a1[a1>0]
    return -np.sum(b*np.log(b))

a=np.array([.1,.3,.4,.2])

# np.random.seed(12)

samples = np.random.dirichlet(10*a,size=100)


print(shannon(a))

vals = [shannon(s) for s in samples ];

np.mean(vals),np.std(vals)
```
