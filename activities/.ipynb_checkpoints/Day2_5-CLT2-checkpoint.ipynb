```r
# =============================================================================
# Activity 9: Central Limit Theorem Revisited
# =============================================================================
# 
# GOAL: Deep understanding of CLT as bridge between probability and inference
#
# CONCEPTS: Sampling distributions, Central Limit Theorem, Standard Error,
#           why inference works
#
# CONNECTION: This directly sets up Day 3 (confidence intervals & hypothesis tests)
# =============================================================================

# -----------------------------------------------------------------------------
# PART 1: Review - Population vs. Sample
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("POPULATIONS AND SAMPLES\n")
cat("=================================================================\n\n")

cat("POPULATION:\n")
cat("• All possible individuals/measurements\n")
cat("• Usually huge or infinite\n")
cat("• Has parameters: μ (mean), σ² (variance)\n")
cat("• We typically DON'T know these!\n\n")

cat("SAMPLE:\n")
cat("• Subset we actually observe\n")
cat("• Size: n\n")
cat("• Has statistics: x̄ (mean), s² (variance)\n")
cat("• We DO know these\n\n")

cat("GOAL OF INFERENCE:\n")
cat("Use sample statistics to learn about population parameters\n\n")


# -----------------------------------------------------------------------------
# PART 2: The Setup - Create a Known Population
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("CREATE A POPULATION (so we can see what happens)\n")
cat("=================================================================\n\n")

set.seed(12345)

# Create a population that is UNIFORM (decidedly NOT normal!)
population <- runif(100000, min = 0, max = 10)

pop_mean <- mean(population)
pop_sd <- sd(population)

cat("Population distribution: UNIFORM (flat, not normal!)\n")
cat("Population mean (μ):", round(pop_mean, 2), "\n")
cat("Population SD (σ):", round(pop_sd, 2), "\n\n")

# Plot population
hist(population,
     breaks = 50,
     col = "lightcoral",
     border = "white",
     main = "Population Distribution (Uniform)",
     xlab = "Value",
     probability = TRUE)
abline(v = pop_mean, col = "red", lwd = 3, lty = 2)


# -----------------------------------------------------------------------------
# PART 3: Take ONE Sample
# -----------------------------------------------------------------------------

cat("\n=================================================================\n")
cat("TAKE ONE SAMPLE\n")
cat("=================================================================\n\n")

# Take a sample of size n = 30
n <- 30
one_sample <- sample(population, size = n)

sample_mean <- mean(one_sample)

cat("Sample size: n =", n, "\n")
cat("Sample mean: x̄ =", round(sample_mean, 2), "\n")
cat("Population mean: μ =", round(pop_mean, 2), "\n")
cat("Difference:", round(sample_mean - pop_mean, 2), "\n\n")

cat("The sample mean is CLOSE to population mean, but not exact!\n")
cat("This is SAMPLING ERROR - unavoidable variability.\n\n")


# -----------------------------------------------------------------------------
# PART 4: The Key Idea - MANY Samples
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("WHAT IF WE TOOK MANY SAMPLES?\n")
cat("=================================================================\n\n")

cat("Imagine 1000 different researchers each took a sample of size", n, "\n")
cat("Each would calculate their own sample mean.\n")
cat("What would the DISTRIBUTION of those means look like?\n\n")

# Simulate 1000 researchers
n_samples <- 1000
sample_means <- numeric(n_samples)

for (i in 1:n_samples) {
  one_sample <- sample(population, size = n)
  sample_means[i] <- mean(one_sample)
}

cat("Created", n_samples, "sample means!\n\n")


# -----------------------------------------------------------------------------
# PART 5: The SAMPLING DISTRIBUTION
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("THE SAMPLING DISTRIBUTION OF THE MEAN\n")
cat("=================================================================\n\n")

# Plot it
hist(sample_means,
     breaks = 30,
     col = "lightblue",
     border = "white",
     main = paste("Distribution of Sample Means (n =", n, ")"),
     xlab = "Sample Mean",
     probability = TRUE)

# Overlay normal curve
x_seq <- seq(min(sample_means), max(sample_means), length.out = 200)
theoretical_normal <- dnorm(x_seq, 
                           mean = pop_mean, 
                           sd = pop_sd / sqrt(n))
lines(x_seq, theoretical_normal, col = "red", lwd = 3)

legend("topright",
       c("Simulated means", "Normal curve"),
       col = c("lightblue", "red"),
       lwd = c(10, 3))

cat("OBSERVATIONS:\n")
cat("1. Distribution is approximately NORMAL!\n")
cat("   (Even though population was UNIFORM)\n")
cat("2. Centers at population mean μ =", round(pop_mean, 2), "\n")
cat("3. Has its own spread (variability)\n\n")

# Calculate properties
mean_of_means <- mean(sample_means)
sd_of_means <- sd(sample_means)

cat("Properties of sampling distribution:\n")
cat("Mean of sample means:", round(mean_of_means, 3), "\n")
cat("Population mean:", round(pop_mean, 3), "\n")
cat("→ Mean of means ≈ μ (unbiased!)\n\n")

cat("SD of sample means:", round(sd_of_means, 3), "\n")
cat("This is called the STANDARD ERROR (SE)\n\n")


# -----------------------------------------------------------------------------
# PART 6: The Standard Error Formula
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("STANDARD ERROR (SE)\n")
cat("=================================================================\n\n")

cat("Standard Error formula: SE = σ / √n\n\n")

theoretical_SE <- pop_sd / sqrt(n)

cat("For our example:\n")
cat("Population SD (σ):", round(pop_sd, 3), "\n")
cat("Sample size (n):", n, "\n")
cat("Theoretical SE:", round(theoretical_SE, 3), "\n")
cat("Observed SE:", round(sd_of_means, 3), "\n\n")

cat("They match! (approximately - simulation variability)\n\n")

cat("KEY INSIGHT: SE tells us how much sample means vary!\n")
cat("Smaller SE → sample means cluster tightly around μ\n")
cat("Larger SE → sample means spread out more\n\n")


# -----------------------------------------------------------------------------
# PART 7: Effect of Sample Size on SE
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("HOW SAMPLE SIZE AFFECTS STANDARD ERROR\n")
cat("=================================================================\n\n")

# Try different sample sizes
sample_sizes <- c(5, 10, 30, 100)
par(mfrow = c(2, 2))

for (n in sample_sizes) {
  # Generate sampling distribution
  sample_means <- replicate(1000, mean(sample(population, size = n)))
  
  # Plot
  hist(sample_means,
       breaks = 30,
       col = "lightgreen",
       border = "white",
       main = paste("n =", n),
       xlab = "Sample Mean",
       probability = TRUE,
       xlim = c(0, 10))
  
  # Overlay theoretical normal
  x_seq <- seq(0, 10, length.out = 200)
  theoretical <- dnorm(x_seq, mean = pop_mean, sd = pop_sd / sqrt(n))
  lines(x_seq, theoretical, col = "red", lwd = 2)
  
  # Show SE
  observed_SE <- sd(sample_means)
  text(2, max(theoretical) * 0.8, 
       paste("SE =", round(observed_SE, 2)),
       col = "red", cex = 1.2)
}

par(mfrow = c(1, 1))

cat("OBSERVATION:\n")
cat("As n increases:\n")
cat("1. Distribution becomes more NARROW (smaller SE)\n")
cat("2. Distribution becomes more NORMAL\n")
cat("3. Sample means cluster more tightly around μ\n\n")

# Quantify the relationship
cat("SE vs. sample size:\n")
for (n in sample_sizes) {
  theoretical_SE <- pop_sd / sqrt(n)
  cat(sprintf("n = %3d:  SE = %.3f\n", n, theoretical_SE))
}

cat("\nSE decreases with √n, not n!\n")
cat("To cut SE in half, need 4× the sample size\n\n")


# -----------------------------------------------------------------------------
# PART 8: The Central Limit Theorem (Formal Statement)
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("THE CENTRAL LIMIT THEOREM\n")
cat("=================================================================\n\n")

cat("FORMAL STATEMENT:\n")
cat("If X₁, X₂, ..., Xₙ are independent random samples from\n")
cat("ANY distribution with mean μ and variance σ²,\n")
cat("then as n → ∞, the sampling distribution of X̄ approaches:\n\n")
cat("    X̄ ~ N(μ, σ²/n)\n\n")
cat("Or equivalently:\n")
cat("    (X̄ - μ) / (σ/√n) ~ N(0, 1)\n\n")

cat("KEY POINTS:\n")
cat("1. Works for ANY original distribution (uniform, exponential, etc.)\n")
cat("2. Larger n → better approximation\n")
cat("3. Rule of thumb: n ≥ 30 usually sufficient\n")
cat("4. If original distribution is already normal, works for any n\n\n")


# -----------------------------------------------------------------------------
# PART 9: CLT with Different Population Distributions
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("CLT WORKS FOR ANY DISTRIBUTION!\n")
cat("=================================================================\n\n")

par(mfrow = c(2, 3))

# 1. Exponential (very skewed)
exp_pop <- rexp(100000, rate = 1)
hist(exp_pop, breaks = 50, main = "Exponential Population", 
     col = "lightcoral", border = "white", xlim = c(0, 10))

exp_means <- replicate(1000, mean(sample(exp_pop, 30)))
hist(exp_means, breaks = 30, main = "Means (n=30)",
     col = "lightblue", border = "white")

# 2. Binomial (discrete)
binom_pop <- rbinom(100000, size = 10, prob = 0.3)
hist(binom_pop, breaks = -0.5:10.5, main = "Binomial Population",
     col = "lightcoral", border = "white")

binom_means <- replicate(1000, mean(sample(binom_pop, 30)))
hist(binom_means, breaks = 30, main = "Means (n=30)",
     col = "lightblue", border = "white")

# 3. Bimodal (two peaks)
bimodal_pop <- c(rnorm(50000, 2, 0.5), rnorm(50000, 8, 0.5))
hist(bimodal_pop, breaks = 50, main = "Bimodal Population",
     col = "lightcoral", border = "white")

bimodal_means <- replicate(1000, mean(sample(bimodal_pop, 30)))
hist(bimodal_means, breaks = 30, main = "Means (n=30)",
     col = "lightblue", border = "white")

par(mfrow = c(1, 1))

cat("AMAZING RESULT:\n")
cat("No matter how weird the population distribution,\n")
cat("the sampling distribution of means is approximately normal!\n\n")


# -----------------------------------------------------------------------------
# PART 10: Connection to Inference (CRITICAL!)
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("WHY THIS MATTERS FOR STATISTICAL INFERENCE\n")
cat("=================================================================\n\n")

cat("SITUATION IN RESEARCH:\n")
cat("• We don't know population mean μ\n")
cat("• We DO observe sample mean x̄\n")
cat("• Question: How far is x̄ from μ?\n\n")

cat("CLT TELLS US:\n")
cat("• Sample means are normally distributed around μ\n")
cat("• With standard error SE = σ/√n\n")
cat("• Therefore, x̄ should be within about 2×SE of μ (95% of the time)\n\n")

cat("THIS LEADS TO:\n")
cat("1. CONFIDENCE INTERVALS (tomorrow!):\n")
cat("   We'll say: μ is probably within x̄ ± 2×SE\n\n")

cat("2. HYPOTHESIS TESTS (tomorrow!):\n")
cat("   We'll ask: If μ were [some value], how likely is our x̄?\n\n")

cat("3. WHY NORMALITY ASSUMPTIONS MATTER:\n")
cat("   We're assuming sampling distribution is normal (CLT)\n")
cat("   Don't need data to be normal, just n large enough!\n\n")


# -----------------------------------------------------------------------------
# PART 11: Practical Example - Blood Pressure
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("PRACTICAL EXAMPLE: BLOOD PRESSURE STUDY\n")
cat("=================================================================\n\n")

# True population (unknown in reality)
true_pop_mean <- 120  # mmHg
true_pop_sd <- 15

cat("SCENARIO:\n")
cat("• Population mean BP: μ = 120 mmHg (but we don't know this!)\n")
cat("• Population SD: σ = 15 mmHg (also unknown)\n")
cat("• We measure n = 25 people\n\n")

# Take ONE sample (like we would in real research)
set.seed(999)
bp_population <- rnorm(100000, mean = true_pop_mean, sd = true_pop_sd)
my_sample <- sample(bp_population, size = 25)
my_sample_mean <- mean(my_sample)

cat("OUR SAMPLE:\n")
cat("• Sample mean: x̄ =", round(my_sample_mean, 1), "mmHg\n")
cat("• This is our ESTIMATE of μ\n\n")

cat("HOW UNCERTAIN ARE WE?\n")
cat("• Standard Error: SE = σ/√n = 15/√25 = 3 mmHg\n")
cat("• So our estimate could easily be off by ±3 mmHg\n")
cat("• 95% of time, x̄ within μ ± 2×SE = μ ± 6\n\n")

# Simulate to verify
many_samples <- replicate(1000, mean(sample(bp_population, size = 25)))

hist(many_samples,
     breaks = 30,
     col = "lavender",
     border = "white",
     main = "Sampling Distribution (n=25)",
     xlab = "Sample Mean BP",
     probability = TRUE)

abline(v = true_pop_mean, col = "red", lwd = 3, lty = 2)
abline(v = true_pop_mean - 2*3, col = "blue", lwd = 2, lty = 2)
abline(v = true_pop_mean + 2*3, col = "blue", lwd = 2, lty = 2)

legend("topright",
       c("True μ", "μ ± 2×SE"),
       col = c("red", "blue"),
       lwd = c(3, 2),
       lty = 2)

cat("TOMORROW: We'll flip this around!\n")
cat("Instead of: 'Where is x̄ given μ?'\n")
cat("We'll ask: 'Where is μ given x̄?'\n")
cat("→ This gives us CONFIDENCE INTERVALS\n\n")


# -----------------------------------------------------------------------------
# PART 12: YOUR TURN - Explore!
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("YOUR TURN: SOLIDIFY YOUR UNDERSTANDING\n")
cat("=================================================================\n\n")

# TODO 1: Change sample size
# Modify the blood pressure example to use n = 100 instead of n = 25
# How does this change the SE?
# How does this change the width of the sampling distribution?


# TODO 2: Very small samples
# Try n = 5 with the uniform population
# Is the sampling distribution still normal?
# How much variability is there?


# TODO 3: Non-normal, small sample
# Population: rexp(100000, rate = 1)
# Sample size: n = 5
# Generate sampling distribution of means
# Is it normal? Try n = 30 instead


# TODO 4: Compare estimators
# Which is a better estimator of μ: mean or median?
# Generate sampling distributions of both
# Which has smaller SE?


# TODO 5: Preview confidence interval
# For the BP example with n = 25:
# Calculate: x̄ ± 2×SE
# Repeat this 100 times (100 different samples)
# How many of the 100 intervals contain true μ?
# Should be about 95!


# -----------------------------------------------------------------------------
# DISCUSSION QUESTIONS
# -----------------------------------------------------------------------------
#
# 1. Why is the CLT called the "Central" Limit Theorem?
#    What makes it so central to statistics?
#
# 2. We showed that sample means are normal even when population isn't.
#    Does this work for other statistics (median, variance, etc.)?
#
# 3. The formula SE = σ/√n means to cut SE in half, we need 4× the data.
#    Why is this a problem in real research? (Think: cost, time, ethics)
#
# 4. In the BP example, we knew σ = 15. In reality, we don't know σ.
#    How would this affect our calculation of SE? (Preview of t-distribution!)
#
# 5. If your population is already perfectly normal, do you still need
#    large n for CLT to work? Why or why not?
#
# 6. Explain to a non-statistician: Why can we use normal-based methods
#    even when our data clearly isn't normally distributed?
#
# =============================================================================
```
