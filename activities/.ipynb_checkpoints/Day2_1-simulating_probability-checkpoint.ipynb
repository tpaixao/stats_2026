```r
# =============================================================================
# Activity 5: Simulating Probability
# =============================================================================
# 
# GOAL: Understand probability through simulation rather than formulas
#
# CONCEPTS: Law of Large Numbers, empirical probability, random sampling
#
# KEY INSIGHT: Simulated frequencies converge to theoretical probabilities
#              as the number of trials increases
# =============================================================================

# -----------------------------------------------------------------------------
# PART 1: Coin Flips - The Simplest Random Process
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("PART 1: COIN FLIPS\n")
cat("=================================================================\n\n")

# Flip a fair coin once
set.seed(2024)
one_flip <- sample(c("H", "T"), size = 1)
cat("One flip:", one_flip, "\n\n")

# Flip 10 times
ten_flips <- sample(c("H", "T"), size = 10, replace = TRUE)
cat("Ten flips:", ten_flips, "\n")
cat("Number of heads:", sum(ten_flips == "H"), "\n")
cat("Proportion heads:", mean(ten_flips == "H"), "\n\n")

# Flip 100 times
hundred_flips <- sample(c("H", "T"), size = 100, replace = TRUE)
cat("100 flips:\n")
cat("Number of heads:", sum(hundred_flips == "H"), "\n")
cat("Proportion heads:", mean(hundred_flips == "H"), "\n\n")

# Flip 10,000 times
many_flips <- sample(c("H", "T"), size = 10000, replace = TRUE)
cat("10,000 flips:\n")
cat("Number of heads:", sum(many_flips == "H"), "\n")
cat("Proportion heads:", mean(many_flips == "H"), "\n\n")

cat("OBSERVATION: As we flip more times, the proportion gets closer to 0.5\n")
cat("This is the LAW OF LARGE NUMBERS\n\n")


# -----------------------------------------------------------------------------
# PART 2: Visualizing Convergence to Probability
# -----------------------------------------------------------------------------

# Let's watch the proportion converge in real-time
n_flips <- 1000
flips <- sample(c("H", "T"), size = n_flips, replace = TRUE)

# Calculate running proportion of heads
cumulative_heads <- cumsum(flips == "H")
cumulative_trials <- 1:n_flips
running_proportion <- cumulative_heads / cumulative_trials

# Plot
plot(1:n_flips, running_proportion, 
     type = "l", col = "blue", lwd = 2,
     xlab = "Number of Flips",
     ylab = "Proportion of Heads",
     main = "Convergence to True Probability (0.5)",
     ylim = c(0, 1))
abline(h = 0.5, col = "red", lwd = 2, lty = 2)
legend("topright", 
       c("Observed proportion", "True probability"),
       col = c("blue", "red"),
       lwd = 2, lty = c(1, 2))

cat("Notice how the proportion bounces around early, then stabilizes near 0.5\n\n")


# -----------------------------------------------------------------------------
# PART 3: Multiple Experiments - Variability
# -----------------------------------------------------------------------------

# What if 10 different people each flip 20 coins?
n_people <- 10
flips_per_person <- 20

people_results <- numeric(n_people)
for (i in 1:n_people) {
  flips <- sample(c("H", "T"), size = flips_per_person, replace = TRUE)
  people_results[i] <- mean(flips == "H")
}

cat("10 people each flip 20 coins:\n")
print(round(people_results, 2))
cat("\nMean of all proportions:", round(mean(people_results), 2), "\n")
cat("SD of proportions:", round(sd(people_results), 2), "\n\n")

# Visualize
hist(people_results, 
     breaks = 10,
     col = "lightblue",
     border = "white",
     main = "Distribution of Proportions\n(10 people, 20 flips each)",
     xlab = "Proportion of Heads")
abline(v = 0.5, col = "red", lwd = 3, lty = 2)

cat("Even though true probability is 0.5, individual results vary!\n")
cat("This is SAMPLING VARIABILITY - we'll use this for inference later.\n\n")


# -----------------------------------------------------------------------------
# PART 4: Dice Rolling - More Complex Probabilities
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("PART 2: ROLLING DICE\n")
cat("=================================================================\n\n")

# Roll one die
one_roll <- sample(1:6, size = 1)
cat("One die roll:", one_roll, "\n\n")

# Roll many times
many_rolls <- sample(1:6, size = 6000, replace = TRUE)

# Count outcomes
cat("Results from 6000 rolls:\n")
table(many_rolls)
cat("\nProportions:\n")
print(round(table(many_rolls) / 6000, 3))

# Visualize
barplot(table(many_rolls),
        col = "lightgreen",
        main = "6000 Die Rolls\n(Each outcome should be ~1/6 = 0.167)",
        xlab = "Die Face",
        ylab = "Frequency")
abline(h = 1000, col = "red", lwd = 2, lty = 2)  # Expected: 6000/6 = 1000


# -----------------------------------------------------------------------------
# PART 5: Sum of Two Dice - A More Interesting Distribution
# -----------------------------------------------------------------------------

cat("\n\n=================================================================\n")
cat("PART 3: SUM OF TWO DICE\n")
cat("=================================================================\n\n")

# When we roll two dice and add them, what's the distribution?
n_rolls <- 10000
sums <- numeric(n_rolls)

for (i in 1:n_rolls) {
  die1 <- sample(1:6, size = 1)
  die2 <- sample(1:6, size = 1)
  sums[i] <- die1 + die2
}

# Or more efficiently:
sums <- replicate(10000, sum(sample(1:6, size = 2, replace = TRUE)))

cat("Sum of two dice - possible values: 2 to 12\n\n")

# Count outcomes
freq_table <- table(sums)
prob_table <- prop.table(freq_table)

cat("Observed frequencies:\n")
print(freq_table)

cat("\nObserved probabilities:\n")
print(round(prob_table, 3))

# Visualize
hist(sums, 
     breaks = seq(1.5, 12.5, by = 1),
     col = "lavender",
     border = "white",
     main = "Sum of Two Dice (10,000 rolls)\nNotice: NOT uniform!",
     xlab = "Sum",
     ylab = "Frequency")

cat("\n\nOBSERVATION: Sum of 7 is most common!\n")
cat("Why? More ways to make 7: (1,6), (2,5), (3,4), (4,3), (5,2), (6,1)\n")
cat("Only one way to make 2: (1,1)\n")
cat("Only one way to make 12: (6,6)\n\n")

# Calculate theoretical probabilities
cat("Theoretical probabilities:\n")
cat("P(sum = 2)  = 1/36 =", round(1/36, 3), " (observed:", round(prob_table["2"], 3), ")\n")
cat("P(sum = 7)  = 6/36 =", round(6/36, 3), " (observed:", round(prob_table["7"], 3), ")\n")
cat("P(sum = 12) = 1/36 =", round(1/36, 3), " (observed:", round(prob_table["12"], 3), ")\n\n")


# -----------------------------------------------------------------------------
# PART 6: Conditional Probability - Given Information
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("PART 4: CONDITIONAL PROBABILITY\n")
cat("=================================================================\n\n")

# Roll two dice 10000 times
n_rolls <- 10000
die1 <- sample(1:6, size = n_rolls, replace = TRUE)
die2 <- sample(1:6, size = n_rolls, replace = TRUE)
sums <- die1 + die2

# Question: What's P(sum = 8)?
p_eight <- mean(sums == 8)
cat("P(sum = 8) =", round(p_eight, 3), "\n\n")

# Question: What's P(sum = 8 | first die = 3)?
# (What's probability sum is 8, GIVEN that first die is 3?)
rolls_with_3 <- sums[die1 == 3]
p_eight_given_3 <- mean(rolls_with_3 == 8)

cat("P(sum = 8 | first die = 3) =", round(p_eight_given_3, 3), "\n")
cat("Makes sense: if first die = 3, second must be 5 for sum = 8\n")
cat("P(second die = 5) = 1/6 =", round(1/6, 3), "\n\n")

# Another question: P(sum > 10)?
p_greater_10 <- mean(sums > 10)
cat("P(sum > 10) =", round(p_greater_10, 3), "\n")
cat("This means sums of 11 or 12\n\n")


# -----------------------------------------------------------------------------
# PART 7: Biology Application - Genetic Crosses
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("PART 5: BIOLOGY APPLICATION - MENDELIAN GENETICS\n")
cat("=================================================================\n\n")

# Cross two heterozygous parents (Aa x Aa)
# Offspring can be: AA (25%), Aa (50%), aa (25%)

simulate_cross <- function() {
  # Each parent gives one allele randomly
  from_parent1 <- sample(c("A", "a"), size = 1)
  from_parent2 <- sample(c("A", "a"), size = 1)
  
  # Offspring genotype
  offspring <- paste0(sort(c(from_parent1, from_parent2)), collapse = "")
  return(offspring)
}

# Simulate 1000 offspring
n_offspring <- 1000
offspring <- replicate(n_offspring, simulate_cross())

cat("Results from 1000 offspring of Aa x Aa cross:\n")
print(table(offspring))

cat("\nProportions:\n")
proportions <- prop.table(table(offspring))
print(round(proportions, 3))

cat("\nExpected (Mendelian ratios):\n")
cat("AA: 0.25 (observed:", round(proportions["AA"], 3), ")\n")
cat("Aa: 0.50 (observed:", round(proportions["Aa"], 3), ")\n")
cat("aa: 0.25 (observed:", round(proportions["aa"], 3), ")\n\n")

# Visualize
barplot(table(offspring),
        col = c("lightblue", "lightgreen", "lightcoral"),
        main = "Genotype Distribution (Aa x Aa)\nExpected 1:2:1 ratio",
        ylab = "Count")
abline(h = c(250, 500), col = "red", lty = 2)


# -----------------------------------------------------------------------------
# PART 8: YOUR TURN - Explore Probabilities
# -----------------------------------------------------------------------------

cat("\n=================================================================\n")
cat("YOUR TURN: EXPLORE!\n")
cat("=================================================================\n\n")

# TODO 1: Biased coin
# What if the coin is unfair: P(H) = 0.7?
# Simulate 1000 flips and verify the proportion is near 0.7

# biased_flips <- sample(c("H", "T"), size=1000, replace=TRUE, prob=c(0.7, 0.3))


# TODO 2: Three dice
# What's the distribution of the sum of THREE dice?
# Simulate and plot


# TODO 3: Birthday problem
# In a room of 23 people, what's the probability at least 2 share a birthday?
# Simulate:
#   - Generate 23 random birthdays (1-365)
#   - Check if any duplicates
#   - Repeat 1000 times
#   - Calculate proportion with matches


# TODO 4: Rare disease screening
# Disease prevalence: 1%
# Test sensitivity: 95% (detects disease if present)
# Test specificity: 90% (negative if no disease)
# Simulate 10000 people:
#   - Assign disease status (1% have it)
#   - Assign test results based on sensitivity/specificity
#   - Calculate: P(have disease | test positive)


# -----------------------------------------------------------------------------
# PART 9: Key Insights About Probability
# -----------------------------------------------------------------------------

cat("\n=================================================================\n")
cat("KEY INSIGHTS:\n")
cat("=================================================================\n\n")

cat("1. LAW OF LARGE NUMBERS:\n")
cat("   - As sample size increases, observed frequency  true probability\n")
cat("   - But small samples can be quite variable!\n\n")

cat("2. PROBABILITY IS A LONG-RUN FREQUENCY:\n")
cat("   - P(H) = 0.5 doesn't mean you'll get exactly 5 heads in 10 flips\n")
cat("   - It means in the long run, ~50% will be heads\n\n")

cat("3. SAMPLING VARIABILITY:\n")
cat("   - Different samples give different results\n")
cat("   - This variability decreases as sample size increases\n")
cat("   - Foundation for statistical inference!\n\n")

cat("4. SIMULATION IS POWERFUL:\n")
cat("   - Can estimate probabilities without complex formulas\n")
cat("   - Useful when analytical solutions are hard\n")
cat("   - Great for checking theoretical calculations\n\n")

cat("5. DISTRIBUTIONS ARISE NATURALLY:\n")
cat("   - Sum of dice: bell-shaped (central limit theorem preview!)\n")
cat("   - Genotypes: multinomial\n")
cat("   - Different processes  different distributions\n\n")


# -----------------------------------------------------------------------------
# PART 10: Connection to Statistical Inference
# -----------------------------------------------------------------------------

cat("=================================================================\n")
cat("CONNECTION TO INFERENCE:\n")
cat("=================================================================\n\n")

cat("What we just did:\n")
cat("1. KNEW the true probability (P(H) = 0.5)\n")
cat("2. Generated data from this probability\n")
cat("3. Observed that sample proportions vary around 0.5\n\n")

cat("What we'll do in statistical inference (Day 3):\n")
cat("1. OBSERVE data (sample)\n")
cat("2. DON'T KNOW true probability\n")
cat("3. Use sample to ESTIMATE probability and quantify uncertainty\n\n")

cat("The sampling variability we saw today is the REASON we need\n")
cat("confidence intervals and hypothesis tests!\n\n")


# -----------------------------------------------------------------------------
# DISCUSSION QUESTIONS
# -----------------------------------------------------------------------------
#
# 1. Why does the proportion of heads bounce around more in early flips
#    than in later flips?
#
# 2. If you flip a fair coin 10 times and get 8 heads, does this prove
#    the coin is biased? Why or why not?
#
# 3. In the two-dice example, why is 7 the most likely sum?
#    Can you list all the ways to get each sum?
#
# 4. We simulated the Mendelian cross Aa x Aa. How would you modify
#    the code for a test cross (Aa x aa)?
#
# 5. How is simulation useful in biology research? Can you think of
#    situations where you can't calculate probability analytically
#    but could simulate it?
#
# 6. The Law of Large Numbers says frequencies converge to probabilities.
#    But how large is "large enough"? Does it depend on the situation?
#
# =============================================================================
```
